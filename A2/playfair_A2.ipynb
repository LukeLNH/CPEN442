```python
# Playfair cipher
def count_bigrams(text):
    bigrams = {}

    for i in range(len(text) - 1):
        if i%2 != 0: continue
        bigram = text[i:i+2]
        if bigram not in bigrams:
            bigrams[bigram] = 0

        bigrams[bigram] += 1
    
    return dict(sorted(bigrams.items(), key=lambda item: item[1], reverse=True))

def count_fourgrams(text):
  bigrams = {}

  for i in range(len(text) - 3):
      if i%2 != 0: continue
      bigram = text[i:i+4]
      if bigram not in bigrams:
          bigrams[bigram] = 0

      bigrams[bigram] += 1

  return dict(sorted(bigrams.items(), key=lambda item: item[1], reverse=True))

def count_sixgrams(text):
  bigrams = {}

  for i in range(len(text) - 5):
      if i%2 != 0: continue
      bigram = text[i:i+6]
      if bigram not in bigrams:
          bigrams[bigram] = 0

      bigrams[bigram] += 1

  return dict(sorted(bigrams.items(), key=lambda item: item[1], reverse=True))
```
